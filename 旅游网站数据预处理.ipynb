# 异常值的处理
- 主要我们处理的字段是”价格“和”节省“两个字段

```python
df.describe()
```

```python
# 转置
df.describe().T
```

```python
## 计算价格的标准差
df['价格'].std()
```

```python
## 计算价格减去价格平均值的差值
df['价格']-df['价格'].mean()
```

```python
## 标准差标准华处理，也叫做标准分数
sta=(df['价格']-df['价格'].mean())/df['价格'].std()
sta
```

```python
## 标准分数>3的我们认为是异常值，现在就把异常值筛选出来
df[abs(sta)>3]
#df[sta.abs()>3]
```

```python
## 请把节省的钱大于价格的钱的数据筛选出来
df[df['节省']>df['价格']]
```

```python
## 数据拼接，把异常值数据拿出来
pd.concat([df[df['节省']>df['价格']],df[abs(sta)>3]])
```

```python
## 从原始数据中删除掉所有的异常值
## 找到异常值的索引
del_index=pd.concat([df[df['节省']>df['价格']],df[abs(sta)>3]]).index
del_index
```

```python
## 执行删除异常数据
df.drop(del_index,inplace=True)
```
