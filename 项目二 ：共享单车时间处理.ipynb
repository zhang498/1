```python
## 时间戳
- 时间戳是1970年1月1日早上8点开始，到现在某一时刻的毫秒值
### 未来要学习mysql,在MySQL中，有时候存储的时间是使用时间戳的形式来进行存储
- 用时间戳的形式来进行存储所占用的硬盘空间会比时间格式所占用的硬盘空间小
- 为了帮企业节省部分硬件资源
```

```python
dtime=data['订单开始时间']
dtime
```

```python
## 把时间转换成时间戳
v=(dtime.values-np.datetime64('1970-01-01 08:00:00'))/np.timedelta64(1,'ms')
v
```

```python
data['订单开始时间']=v
data.head()
```

```python
data.info()
```

```python
# 把时间戳再转换为时间
data['订单开始时间']= pd.to_datetime(data['订单开始时间'],origin=pd.Timestamp('1970-01-01 08:00:00'),unit='ms')
data.head()
```
