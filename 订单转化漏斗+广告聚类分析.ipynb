```python
## 区域分布地图
from pyecharts.charts import Map
# 为了保持下边的地理分布图使用的省份一致，我们需要处理一下数据
def strip_region(iterable):
    result=[]
    for i in iterable:
        if i.endswith('自治区'):
            if i=='内蒙古自治区':
                i==i[:3]
                result.append(i)
            else:
                result.append(i[:2])
        else:
            result.append(i)
    return result       
```

```python
## 处理自治区
data_address.index=strip_region(data_address.index)
## 处理省,把末尾的省字去掉
data_address.index=data_address.index.str.strip('省')
data_address
```

```python
## 把数据编程地图要求的格式
data_map=[list(z) for z in data_address.items()]
```

```python
m=(
    Map()
    .add('订单分布地图',data_map,'china')    
)
m.render_notebook()
```

## 订单各环节转换漏斗图


```python
## 总订单数
total_order_count=data['订单编号'].size
total_order_count
```

```python
## 付款订单数
pay_order_count=data['订单付款时间'].notnull().sum()
pay_order_count
```

```python
## 到款订单数
completed_deal_count=(data['买家实际支付金额']!=0.0).sum()
completed_deal_count
```

```python
## 全额到款订单数
full_pay_order_count=(data['退款金额']==0.0).sum()-data['订单付款时间'].isnull().sum()
full_pay_order_count
```

```python
order_count_data=pd.Series(
    {
        "总订单数":total_order_count,
        "付款订单数":pay_order_count,
        "到款订单数":completed_deal_count,
        "全额到款订单数":full_pay_order_count
    },name='订单数'
).to_frame()
order_count_data
```

```python
from pyecharts.charts import Funnel
name='订单数'
f= Funnel().add(series_name=name,
                data_pair=[list(z) for z in zip(order_count_data.index,order_count_data[name])]
               ).set_global_opts(title_opts=opts.TitleOpts(title='订单转化'))
f.render_notebook()
```

```python
from sklearn.cluster import KMeans
x1=data[['日均UV','平均注册率','平均搜索量','访问深度','订单转化率','投放总时间']]
inertia=[]

for i in range(1,11):
    km=KMeans(n_clusters=i)
    km.fit(x1)
    inertia.append(km.inertia_)
l=(
    Line(init_opts=opts.InitOpts(width='1000px',height='500px'))
    .add_xaxis(range(1,11))
    .add_yaxis('',inertia)
)
l.render_notebook()
```

```python
km=KMeans(n_clusters=5)
y_means=km.fit_predict(x1)
y_means
```

```python
## 柱状图与折线图叠加显示
x_axis=list(data['渠道代号'].head(10).values)
y1_axis=list((data['平均注册率']*100).head(10))
y2_axis=list((data['订单转化率']*100).head(10))
y3_axis=list((data['平均搜索量']*100).head(10))
```

```python
bar=(
    Bar()
    .add_xaxis(x_axis)
    .add_yaxis('平均注册率',y1_axis)
    .add_yaxis('订单转化率',y2_axis)
    .extend_axis(
        yaxis=opts.AxisOpts(
            axislabel_opts=opts.LabelOpts(formatter='{value}个')
        )
    )
    .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
    .set_global_opts(
        title_opts=opts.TitleOpts('折线图与柱状图叠加'),
        yaxis_opts=opts.AxisOpts(axislabel_opts=opts.LabelOpts(formatter='{value}%')),
        datazoom_opts=[
            opts.DataZoomOpts(yaxis_index=0),
            opts.DataZoomOpts(type_='inside',yaxis_index=0)            
        ]
    )
)
line=Line().add_xaxis(x_axis).add_yaxis('搜索量',y3_axis,yaxis_index=1)
bar.overlap(line)
bar.render_notebook()
```
